```python
def get_pic(category, count):
    for i in range(1, 100):
        pic_list = []
        pic_url = "http://wallpaper.apc.360.cn/index.php?c=WallPaper&a=getAppsByCategory&cid=%s&start=%s&count=%s&from=360chrome" % (category, str(i), count)
        pic = requests.get(pic_url)
        pic_data = pic.json()["data"]
        if pic_data:
            tmp = deal_pic_data(pic_data)
        else:
            break
        time.sleep(5)
```

```python
def download_img(img_url, name):
    print (img_url)
    r = requests.get(img_url, stream=True)
    print(r.status_code) # 返回状态码
    if r.status_code == 200:
        open("D:\\LinkSpace\\Pictures\\Saved Pictures\\img\\" + name + '_img.png', 'wb').write(r.content) # 将内容写入图片
        print("done")
    del r
```

```python
def save_tag(tag, name):
    print(tag)
    with open("tag\\" + name + ".txt", "w") as f:
        f.write(tag)
```

```python
def deal_pic_data(data):
    my_list = []
    num = 0
    for i in data:
        num += 1
        tmp = [i["url"], i["tag"]]
        download_img(i["url"], str(num))
        # save_tag(i["tag"], str(num))
        my_list.append(tmp)
    return my_list
```
